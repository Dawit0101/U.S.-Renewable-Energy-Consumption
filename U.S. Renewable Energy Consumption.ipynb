```python
#import Library  
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import datasets 
df = pd.read_csv('dataset.csv')
df.head(10)
```

```python
# Replace 0 values with NaN
df.replace(0, float('nan'), inplace=True)
```

```python
# Total renewable energy consumption over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Year', y='Total Renewable Energy', marker='o')
plt.title('Total Renewable Energy Consumption Over Time')
plt.xlabel('Year')
plt.ylabel('Total Renewable Energy (trillion BTUs)')
plt.grid(True)
plt.show()
```

```python
# Sector-wise comparison
plt.figure(figsize=(12, 8))
sns.barplot(data=df, x='Sector', y='Total Renewable Energy', estimator=sum)
plt.title('Total Renewable Energy Consumption by Sector')
plt.xlabel('Sector')
plt.ylabel('Total Renewable Energy (trillion BTUs)')
plt.xticks(rotation=45)
plt.show()
```

```python
#Comparative Analysis
#Group the data by 'sector' and calculate the total renewable energy consumption for each sector and each energy source
sector_source_df = df.groupby('Sector').agg({
    'Hydroelectric Power': 'sum',
    'Geothermal Energy': 'sum',
    'Solar Energy': 'sum',
    'Wind Energy': 'sum',
    'Wood Energy': 'sum',
    'Waste Energy': 'sum',
    'Fuel Ethanol, Excluding Denaturant': 'sum',
    'Biomass Losses and Co-products': 'sum',
    'Renewable Diesel Fuel': 'sum',
    'Other Biofuels': 'sum',
    'Conventional Hydroelectric Power': 'sum',
    'Biodiesel': 'sum'
}).reset_index()

# Melt the DataFrame to convert it from wide to long format for easier plotting
sector_source_melted = sector_source_df.melt(id_vars='Sector', var_name='Energy Source', value_name='Total Energy Consumption (trillion BTUs)')

# Plot the comparison using a barplot
plt.figure(figsize=(14, 8))
sns.barplot(data=sector_source_melted, x='Sector', y='Total Energy Consumption (trillion BTUs)', hue='Energy Source')
plt.title('Renewable Energy Consumption by Sector and Source')
plt.xlabel('Sector')
plt.ylabel('Total Renewable Energy Consumption (trillion BTUs)')
plt.xticks(rotation=45)
plt.legend(title='Energy Source', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Forecasting using Moving Average with Plotly

import plotly.graph_objects as go

# Calculate the moving average
window_size = 12  # You can adjust the window size as needed
rolling_avg = df['Total Renewable Energy'].rolling(window=window_size).mean()

# Create a Plotly figure
fig = go.Figure()

# Add traces for actual data and moving average forecast
fig.add_trace(go.Scatter(x=df['Year'], y=df['Total Renewable Energy'], mode='lines', name='Actual'))
fig.add_trace(go.Scatter(x=df['Year'], y=rolling_avg, mode='lines', name=f'Moving Average (Window Size={window_size})'))

# Update layout
fig.update_layout(title='Forecast of Total Renewable Energy Consumption using Moving Average',
                  xaxis_title='Year',
                  yaxis_title='Total Renewable Energy (trillion BTUs)',
                  legend=dict(yanchor="top", y=0.99, xanchor="left", x=0.01),
                  template='plotly_white')

# Show plot
fig.show()
```

```python
#policy timeline data

policy_data = pd.DataFrame({
    'Year': [2005, 2010, 2015, 2020],
    'Policy': ['Renewable Energy Tax Credits Introduced',
               'Renewable Portfolio Standards Implemented',
               'Investment in Renewable Infrastructure',
               'Carbon Pricing Initiative'],
    'Description': ['Federal tax credits for renewable energy production were introduced.',
                    'States began implementing renewable portfolio standards to increase renewable energy usage.',
                    'Government allocated funds for renewable energy infrastructure projects.',
                    'Carbon pricing initiative aimed at incentivizing emissions reduction and renewable energy adoption.']
})

# Merge policy data with renewable energy dataset
df_merged = pd.merge(df, policy_data, on='Year', how='left')

# Visualize policy timeline overlaying renewable energy consumption trends
plt.figure(figsize=(12, 8))
sns.lineplot(data=df, x='Year', y='Total Renewable Energy', marker='o', label='Renewable Energy Consumption')
for i, policy in enumerate(policy_data['Policy']):
    plt.axvline(x=policy_data['Year'][i], color='r', linestyle='--', label=policy, linewidth=1)
plt.title('Renewable Energy Consumption with Policy Timeline')
plt.xlabel('Year')
plt.ylabel('Total Renewable Energy (trillion BTUs)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Visualization and Reporting

# Visualize trends in renewable energy consumption over time
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Total Renewable Energy'], marker='o', color='blue', linestyle='-')
plt.title('Trends in Renewable Energy Consumption')
plt.xlabel('Year')
plt.ylabel('Total Renewable Energy (trillion BTUs)')
plt.grid(True)
plt.show()

# Generate summary statistics
summary_statistics = df.describe()

# Print summary statistics
print("Summary Statistics:")
print(summary_statistics)

# Save summary statistics to a file
summary_statistics.to_csv('summary_statistics.csv')
print("Summary statistics saved to 'summary_statistics.csv'.")
```
